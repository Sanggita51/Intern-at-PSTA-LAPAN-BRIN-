```python
import seaborn as sns
```

```python
import pandas as pd
data = pd.read_excel('Seluruh tanggal.xlsx', sheet_name='Plot JKT')
data
```

```python
baru =[]
for i in data['TBB < 250']:
    if i>=0 and i<=25:
        baru.append('Few')
    elif i>25 and i<=50:
        baru.append('Scattered')
    elif i>50 and i<=87:
        baru.append('Broken')
    else:
        baru.append('Overcast')
baru
```

```python
#del(data['Awan'])
data['Tutupan Awan']=baru
data
```

```python
import matplotlib.pyplot as plt
#fig, ax = plt.subplots(figsize=(5, 20))
#ax = sns.barplot(x='ISPU', y='STRD', data=data, ci="sd", capsize=.2, palette="Blues")
ax = sns.barplot(x='Tutupan Awan', y='STR', data=data,capsize=.2, palette="Blues")
ax.set_title("Jakarta Barat")
ticks = [-50,-45,-40,-35,-30,-25,-20,-15,-10,-5,0] #STR
#ticks = [386, 390, 394, 398,402,406,410,414,418,422,426,430] #STRD
ax.set_yticks(ticks)
#ax.set(ylim=(385, 432))
ax.set(ylim=(0, -50))
```
