```python
import seaborn as sns
```

```python
import pandas as pd
data = pd.read_excel('Seluruh tanggal.xlsx', sheet_name='Plot JKT')
data
```

```python
from scipy.stats import pearsonr #jika berdistribusi normal
#from scipy.stats import spearmanr #jika tidak berdistribusi normal
import numpy as np

corr = pearsonr(data[''], data['STRD'])
corr = [np.round(c, 2) for c in corr]
print(corr)
```

```python
import matplotlib.pyplot as plt
fig = plt.gcf()

# Change seaborn plot size
fig.set_size_inches(10, 6)

# Extracting the r-value and the p-value:
#text = 'r=%s, p=%s' % (corr[0], corr[1])
text = 'r=%s, p<0.05' % (corr[0])

# Creating the regression plot:
ax = sns.regplot(x="STR", y="STRD", data=data)
ax.set_yticklabels(ax.get_yticks(), size = 15)
ax.set_xticklabels(ax.get_xticks(), size = 15)
ax.set_xlabel("STR (Watt/m2)", fontsize = 15)
ax.set_ylabel("STRD (Watt/m2)", fontsize = 15)
ax.set_title("Jakarta Barat", fontsize=18)

# Adding the text to the Seaborn plot:
ax.text(-54, 433, text, fontsize=18)
```
