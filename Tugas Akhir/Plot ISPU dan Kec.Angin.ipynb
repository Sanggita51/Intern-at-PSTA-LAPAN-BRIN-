```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
data = pd.read_excel('Seluruh tanggal.xlsx', sheet_name='Plot JKT')
data
```

```python
sumbu_x=[" 18:00", " 19:00", " 20.00", " 21.00", " 22.00", " 23.00", "00.00", "01.00", "02.00", "03.00", "04.00", "05.00"]
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(8, 4))
# make a plot

ax = sns.lineplot(sumbu_x, data[60:72].PM25, color="red", marker="o")
# set x-axis label
ax.set_xlabel("Waktu",fontsize=16)
# set y-axis label
ax.set_ylabel("PM2.5 (μg m-3)",color="red",fontsize=16)
ticks = [0,10,20,30,40,50,60]
ax.set_yticks(ticks)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
#make a plot with different y-axis using second axis object
ax2.plot(sumbu_x, data[300:312].Angin,color="blue",marker="o")
ax2.set_ylabel("Kec.Angin (m/s)",color="blue",fontsize=16)
ticks1 = [0,1,2,3,4,5]
ax2.set_yticks(ticks1)
ax.set_title("Tanggal 6", fontsize = 16)

plt.show()
```
